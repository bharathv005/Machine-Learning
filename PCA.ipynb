```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
dataset = load_iris()
```

```python
X = dataset.data
y = dataset.target
names = dataset.target_names
```

```python
from sklearn.decomposition import PCA
model = PCA(n_components=2) #Number of components to keep
y_means = model.fit(X).transform(X)
```

```python
plt.figure()
colors = ['red','blue','green']
for color,i, target in zip(colors,[0,1,2],names):
    plt.scatter(y_means[y==i,0],y_means[y==i,1],color=color,lw = 2,label = target)
plt.title('IRIS Clusterring')
plt.show()
```
