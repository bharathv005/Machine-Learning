```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("dataset.csv")
```

```python
print(dataset.shape)
print(dataset.head(5))
```

```python
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(dataset.area,dataset.price,color='blue',marker='o')
```

```python
x = dataset.drop('price',axis='columns')
y=dataset.price
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
x=65432
landarea = [[x]]
pred=model.predict(landarea)
print(pred)
```

```python
m = model.coef_
b = model.intercept_
y = m*x+b
print('The price of {0} sqft land is {1}'.format(x,y[0]))
```
